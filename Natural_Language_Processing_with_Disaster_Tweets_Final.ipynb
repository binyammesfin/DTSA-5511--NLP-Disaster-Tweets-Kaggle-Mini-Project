### **3.2 Data Cleaning and Preprocessing**

Text preprocessing is crucial in preparing raw text data for analysis and modeling in natural language processing (NLP) tasks. The preprocessing function provided below performs several important operations on the text data. First, it converts all text to lowercase, ensuring uniformity and reducing the complexity of the text data by treating "Disaster" and "disaster" as the same word. Next, it removes URLs, mentions, and hashtags, which are often irrelevant to the content of the text and can introduce noise into the model. Special characters and numbers are also removed to focus solely on the linguistic content of the tweets.

The text is then tokenized, breaking it down into individual words, which are easier to analyze and model. Stopwords, which are common words that typically do not carry significant meaning (such as "the", "and", "is"), are removed to reduce the dimensionality of the data and help the model focus on more informative words. Finally, lemmatization is performed, reducing words to their base or root form (e.g., "running" to "run"), which helps in treating different forms of a word as a single entity and enhances the consistency of the text data.

**Note:** Please downloading nltk.download('wordnet') as it is necessary to access the Punkt tokenizer, which is required for accurately tokenizing the text into individual words and sentences. I did not include it in the code to avoid multiple downloads.

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
## Text preprocessing function
def preprocess_text(text):
    text = text.lower()  # Lowercase text
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)  # Remove URLs
    text = re.sub(r'\@\w+|\#','', text)  # Remove mentions and hashtags
    text = re.sub(r'[^A-Za-z0-9]+', ' ', text)  # Remove special characters and numbers
    tokens = word_tokenize(text)  # Tokenize text
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]  # Remove stopwords
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]  # Lemmatize tokens
    return ' '.join(tokens)

train['cleaned_text'] = train['text'].apply(preprocess_text)
test['cleaned_text'] = test['text'].apply(preprocess_text)
```

##### Model Architecture and Reasoning

We started with a basic model architecture, specifically Logistic Regression, as it provides a simple and interpretable baseline. Logistic Regression is effective for binary classification tasks and helps establish a benchmark performance level. It transforms text data into numerical features using TF-IDF, allowing the model to weigh the importance of different words and make predictions. Despite its simplicity, Logistic Regression often performs well in text classification tasks due to its linear decision boundaries and ease of implementation.

Next, we included intermediate models such as Support Vector Machine (SVM) and Random Forest. SVM is known for its effectiveness in high-dimensional spaces and robustness to overfitting, making it a good candidate for text data, which often has high dimensionality due to the large vocabulary. Random Forest, an ensemble learning method, captures non-linear relationships and interactions between features by combining multiple decision trees. Both models, while more complex than Logistic Regression, did not significantly outperform the baseline in this case, highlighting the need for more sophisticated approaches.

To further enhance performance, we turned to advanced models, including Recurrent Neural Networks (RNN), Long Short-Term Memory networks (LSTM), and Bidirectional LSTMs. These architectures are specifically designed to handle sequential data and capture temporal dependencies, making them well-suited for text classification. RNNs process sequences in order, allowing them to retain context, while LSTMs address the vanishing gradient problem, enabling the capture of longer dependencies. Bidirectional LSTMs further improve on this by processing sequences in both forward and backward directions, capturing context from both ends. Although these advanced models initially underperformed, their theoretical capability to model complex patterns and dependencies makes them promising candidates for further optimization through hyperparameter tuning. This potential for higher performance justifies their inclusion and continued refinement in our modeling approach.

### **5.1 Hyperparameter Tuning for Bidirectional LSTM**

```python
from tqdm import tqdm
from sklearn.model_selection import ParameterGrid
# Tokenization and padding for Bidirectional LSTM
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(train['cleaned_text'])
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_val_seq = tokenizer.texts_to_sequences(X_val)
test_seq = tokenizer.texts_to_sequences(test['cleaned_text'])

maxlen = 100
X_train_pad = pad_sequences(X_train_seq, maxlen=maxlen)
X_val_pad = pad_sequences(X_val_seq, maxlen=maxlen)
X_test_pad = pad_sequences(test_seq, maxlen=maxlen)

class KerasModelWrapper(BaseEstimator, ClassifierMixin):
    def __init__(self, learning_rate=0.001, dropout_rate=0.5, num_units=128, batch_size=32, epochs=5):
        self.learning_rate = learning_rate
        self.dropout_rate = dropout_rate
        self.num_units = num_units
        self.batch_size = batch_size
        self.epochs = epochs
        self.model = None

    def build_model(self):
        model = Sequential([
            Embedding(input_dim=10000, output_dim=128, input_length=maxlen),
            Bidirectional(LSTM(self.num_units)),
            Dropout(self.dropout_rate),
            Dense(1, activation='sigmoid')
        ])
        model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=self.learning_rate),
                      loss='binary_crossentropy',
                      metrics=['accuracy'])
        return model

    def fit(self, X, y):
        self.model = self.build_model()
        self.model.fit(X, y, epochs=self.epochs, batch_size=self.batch_size, verbose=0)
        return self

    def predict(self, X):
        return (self.model.predict(X) > 0.5).astype("int32")

# Define the grid of hyperparameters to search (I had to be conservative to save computing time)
param_grid = {
    'learning_rate': [0.001, 0.01, 0.1],
    'dropout_rate': [0.3, 0.5],
    'num_units': [64, 128],
    'batch_size': [32, 64],
    'epochs': [5]
}

# Create a parameter grid
param_combinations = list(ParameterGrid(param_grid))
total_combinations = len(param_combinations)

# Progress tracking with tqdm
pbar = tqdm(total=total_combinations)

# Store results
best_f1_score = 0
best_params = None

# Perform Grid Search manually
for params in param_combinations:
    model_wrapper = KerasModelWrapper(**params)
    model_wrapper.fit(X_train_pad, y_train)
    y_val_pred = model_wrapper.predict(X_val_pad)
    f1 = f1_score(y_val, y_val_pred)

    if f1 > best_f1_score:
        best_f1_score = f1
        best_params = params

    pbar.update(1)

# Close the progress bar
pbar.close()
```

```python
print(f'Best Hyperparameters: {best_params}')
print(f'Best F1 Score: {best_f1_score}')
```

```python
# Retrain the best model on the entire training set using the best hyperparameters
best_model = KerasModelWrapper(**best_params)
best_model.fit(X_train_pad, y_train)
```

```python
# Make predictions on the test set
test_predictions = best_model.predict(X_test_pad)
```

```python
# Create a submission file
submission = pd.DataFrame({'id': test['id'], 'target': test_predictions.flatten()})
submission['target'] = submission['target'].astype(int)  # Ensure the target is integer type
submission.to_csv('submission.csv', index=False)
```
